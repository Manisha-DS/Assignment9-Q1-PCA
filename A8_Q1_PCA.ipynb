**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
import warnings
warnings.filterwarnings ('ignore')
```

**Load Dataset**

```python
df=pd.read_csv("wine.csv")
df
```

# EDA

```python
df.info()
```

```python
df1=df.iloc[:,1:]
df1
```

```python
# Converting into numpy array
DF=df1.values
DF
```

```python
# Normalizing the numerical data
DF_N=scale(DF)
DF_N
```

```python
# Transforming to PCA values
DF_PCA=PCA(n_components=13)
DF_PCA_values=DF_PCA.fit_transform(DF_N)
DF_PCA_values
```

```python
# Loadings / Weights
DF_PCA.components_
```

```python
# The amount of variance that each PCA explains is
DF_var=DF_PCA.explained_variance_ratio_
DF_var
```

```python
# Cumulative variance
DF_Cvar=np.cumsum(np.round(DF_var,decimals = 4)*100)
DF_Cvar
```

# Plot For Variance

```python
# Variance plot for PCA components obtained 
plt.plot(DF_Cvar,color="red")
```

# Final DataFrame

```python
Final_DF=pd.concat([df['Type'],pd.DataFrame(DF_PCA_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
Final_DF
```

# Visualization Of PC's

```python
fig=plt.figure(figsize=(12,12))
sb.scatterplot(data=Final_DF)
```

# Cluster Algorithms

**1) H-Clustering**

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# We already have Normalized Data so, we create Dendogram
plt.figure(figsize=(12,12))
dendrogram=sch.dendrogram(sch.linkage(DF_N,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(DF_N),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
df2=df.copy()
df2['clustersid']=hclusters.labels_
df2
```

**2) KMeans Clustering**

```python
from sklearn.cluster import KMeans
```

```python
# As we aready have normalized data
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
# random state can be anything from 0 to 42, but the same number to be used everytime,so that the results don't change
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(DF_N)
    wcss.append(kmeans.inertia_)
```

# Elbow Graph

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
